# YouTube AdView Prediction

Predicts the number of ad views a YouTube video will receive based on engagement metrics.

**Models benchmarked:** Linear Regression · Decision Tree · Random Forest · SVR · ANN

> All logic lives in `src/`. This notebook is an interactive walkthrough — run `python3 src/train.py` for the full pipeline without Jupyter.

```python
import sys
sys.path.insert(0, '..')
import pandas as pd
from src.preprocess import load_and_prepare, clean
from src.config import TRAIN_CSV, TEST_CSV
```

## 1. Load and Explore Training Data

```python
df_raw = pd.read_csv(f'../{TRAIN_CSV}')
print(df_raw.shape)
df_raw.head()
```

```python
df_raw.info()
```

## 2. EDA

```python
from src.eda import plot_category_distribution, plot_adview_distribution, plot_correlation_heatmap
from src.preprocess import clean

df_clean = clean(df_raw, is_train=True)

plot_category_distribution(df_clean)
plot_adview_distribution(df_clean)
plot_correlation_heatmap(df_clean)
```

## 3. Preprocess and Split

```python
X_train, X_test, y_train, y_test, scaler = load_and_prepare(
    f'../{TRAIN_CSV}', is_train=True, save_scaler=True
)
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
```

## 4. Train and Evaluate All Models

```python
from src.train import (
    train_linear, train_decision_tree, train_random_forest, train_svr, train_ann, evaluate
)

lr  = train_linear(X_train, y_train)
dt  = train_decision_tree(X_train, y_train)
rf  = train_random_forest(X_train, y_train)
svr = train_svr(X_train, y_train)
ann = train_ann(X_train, y_train, n_features=X_train.shape[1])
```

```python
import pandas as pd

results = [
    evaluate(lr,  X_test, y_test, 'Linear Regression'),
    evaluate(dt,  X_test, y_test, 'Decision Tree'),
    evaluate(rf,  X_test, y_test, 'Random Forest'),
    evaluate(svr, X_test, y_test, 'SVR'),
    evaluate(ann, X_test, y_test, 'ANN'),
]

pd.DataFrame(results).set_index('model')
```
